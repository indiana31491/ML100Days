```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

#畫出完整的 cos 圖形
x = np.arange(0,360)
y = np.cos(x * np.pi / 180)

plt.plot(x,y)

# 如果要存成圖形檔:
# 把 pyplot.show() 換成下面這行:
plt.savefig("cos函數.png",dpi=300,format="png")
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.scatter(X, Y, c = np.cos(Y/X) )

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
