```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def practice(x):
    return np.cosh(x) + x**2

x = np.arange(-5, 5, 0.01)

plt.plot(x, practice(x))
plt.show()
```

```python
min_practice = optimize.minimize_scalar(practice)
min_practice.x
```
