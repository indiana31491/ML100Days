```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Seaborn圖樣式
有五種預設seaborn主題：darkgrid，whitegrid，dark，white，和ticks。它們各自適合不同的應用和個人喜好。默認主題是darkgrid

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python

# 當刻度不覆蓋整個軸範圍時，該trim參數將限制脊柱的範圍。

f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True);
```

```python
# 使用其他參數控制刪除哪些despine()

sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

# 暫時設定圖形樣式
雖然來回切換很容易，但您也可以axes_style()在with語句中使用該功能來臨時設置繪圖參數。
這也允許您使用不同風格的軸製作圖形

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
f = plt.figure()
with sns.axes_style("darkgrid"):
    ax = f.add_subplot(1, 2, 1)
    sinplot()
ax = f.add_subplot(1, 2, 2)
sinplot(-1)
```

統計關係
統計分析是一個了解數據集中變數之間如何相互關聯以及這些關係如何依賴於其他變數的過程。

將變數與散點圖關聯
散點圖是統計可視化的支柱。它使用點雲描繪了兩個變數的聯合分佈,其中每個點表示數據集中的觀測值。這種描述允許眼睛推斷出大量有關它們之間是否有任何有意義的關係的資訊。

```python
sns.set(style="ticks")
tips = sns.load_dataset("tips")
```

```python
# 散點圖關聯變量
sns.set(style="darkgrid")
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
#當點以兩個維度繪製時,可以通過根據第三個變數對點進行著色,將另一個維度添加到繪圖中。這稱為使用「色調語義」,因為點的顏色獲得意義:
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
# 通過獨立地改變每個點的色調和样式，也可以表示四個變量
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
#聚合和表示不確定性 -- 
#更複雜的數據集將具有變數相同值的多個度量。Seaborn 中的預設行為是透過繪製平均值和 95% 信心區間圍繞平均值來聚合每個值的多個測量值:xx
# kind 圖形類別 - 
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
#信心區間是使用引導計算的,對於較大的數據集,這需要時間。
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```python
#另一個不錯的選擇(尤其是對於較大的數據)是通過繪製標準差而不是信心區間來表示每個時間點的分佈分佈:
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

使用語義映射繪製數據子集。
lineplot()具有與scatterplot()相同的靈活性:它可以通過修改繪圖元素的色調、大小和樣式來顯示最多三個附加變數。

在lineplot()中使用語義也將確定數據的聚合方法。例如,添加具有兩個級別的色調語義將繪圖拆分為兩行和誤差帶,對每行進行著色,以指示它們對應的數據子集。

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

#載入範例資料集 long-form data
fmri = sns.load_dataset("fmri")

# 繪製不同事件與對應數據的分布結果 
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)

#x, y: 特徵 , data: 繪圖資料集
#hue & style:取決於是否推斷出該變量代表“數值”或“分類”數據。默認情況下，數字變量用顏色圖示，圖例條目顯示規則的“刻度”，
```
